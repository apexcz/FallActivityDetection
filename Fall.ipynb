```python
import pandas as pd
import matplotlib.pyplot as plt

'''
FALL ACTIVITY
From the scatter plot graph generated from the data,
it is observed that there is a great deviation of the x,y,z accelerometer readings
'''
def extractTime(occurence):
    time_split = occurence.split(' ')[1]
    hours_minute = time_split[:8]
    return hours_minute

activity_file = 'timeLine.csv'
df = pd.read_csv(activity_file)
df['time'] = df.time.apply(extractTime)


actor_A_wrist = df.loc[df.uid == 'a0:e6:f8:00:00:c0']

fall = actor_A_wrist.loc[actor_A_wrist.activity == 'Fall']
fig = plt.figure()
plt.scatter(list(fall.time.values),list(fall.x.values),color='r',label='x')
plt.scatter(list(fall.time.values),list(fall.y.values),color='g',label='y')
plt.scatter(list(fall.time.values),list(fall.z.values),color='b',label='z')

plt.legend(loc='upper left')
plt.title('Fall activity from the wrist wearable of an actor')
plt.xlabel('Time')
plt.ylabel('Accelerometer reading')
plt.gcf().autofmt_xdate()

fig.savefig('fall.png')
plt.show()
```
