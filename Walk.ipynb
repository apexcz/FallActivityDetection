```python
import pandas as pd
import matplotlib.pyplot as plt

'''
Walk ACTIVITY
From the scatter plot graph generated from the data,
it is observed that the x,y,z accelerometer readings are closely located,
meaning that that there is coordination in motion 
'''
def extractTime(occurence):
    time_split = occurence.split(' ')[1]
    hours_minute = time_split[:8]
    return hours_minute

activity_file = 'timeLine.csv'
df = pd.read_csv(activity_file)
df['time'] = df.time.apply(extractTime)


actor_A_wrist = df.loc[df.uid == 'a0:e6:f8:00:00:c0']

walk_flat_ground = actor_A_wrist.loc[actor_A_wrist.activity == 'Walk on flat ground']

fig = plt.figure()
plt.scatter(list(walk_flat_ground.time.values),list(walk_flat_ground.x.values),color='r',label='x')
plt.scatter(list(walk_flat_ground.time.values),list(walk_flat_ground.y.values),color='g',label='y')
plt.scatter(list(walk_flat_ground.time.values),list(walk_flat_ground.z.values),color='b',label='z')

plt.legend(loc='upper left')
plt.title('Walk activity from the wrist wearable of an actor')
plt.xlabel('Time')
plt.ylabel('Accelerometer reading')
plt.gcf().autofmt_xdate()
fig.savefig('walk.png')
plt.show()
```
